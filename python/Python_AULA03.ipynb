## Desvio Condicional if, else, elif

O desvio condicional é invocado pelo comando **if** na sua forma mais simples

```
if <condição>:
  comandos
```
A *condição* não é nada mais que uma expressão que retorne um valor booleano (ou lógico) True or False.

Se a expressão condicional vier a ser "True", o bloco interno será executado.

Observe que o comando if, tem um ":" no final da linha. Isso significa que a proxima instrução será um comando indentado ou um bloco de comandos.

Vamos ver um exemplo simples.


````
# vamos ordernar alguns números, toda vez que o numero em "a" for maior que "b"
# o programa ajusta em ordem (menor pro maior)

a=int(input('primeiro número:'))
b=int(input(' segundo número:'))

if (a>b):
  # troco os valores de posição
  temp=a # armazeno o valor de a em algum lugar temporario
  a=b    # atribuo a "a" o valor de "b"
  b=temp # atribuo a "b" o valor salvo de "a"

print(a,b)
````


### if else

Há situações que precisamos tomar uma decisão binária. Se a condição não for satisfeita (Verdadeira) tem que ser feita de outra forma. A sintaxe é assim:

```
if (condição):
   comandos caso verdadeiro
else:
   comandos caso falso
```


````
# A eleição está aí então somos ou não somos obrigados a votar?

idade=int(input("Qual a sua idade?"))

if( 18<=idade<=70 ):
  print("Você é obrigado a votar")
else:
  print("Você nao precisa votar")
````


### if/elif/else

O Python não tem comandos especializados em fazer desvios baseado em resultados de expressões. Como se fosse um indice. A técnica usada nesse caso é aninhar os ifs

```
if x==1:
  comando se x for 1
else:
  if x==2:
    comando se x for 2
  else:
    if x==3:
      comandd se x for 3
    else:
      comandos se for qualquer outro valor
```
Temos que tomar cuidado que na técnica acima, a identação é fácil de se perder.

Pelo conceito de legibilidade do python, esse código não é bem legível e pode ser simplificado juntando as cláusulas **else: if** em uma única cláusula **elif**

```
if x==1:
  comando se x for 1
elif x==2:
  comando se x for 2
elif x==3:
  comandd se x for 3
else:
  comandos se for qualquer outro valor
```

A sequencia if/elif/else tem uma particularidade. Apesar do elif e else serem opcionais. Se ambos tiverem que aparecer, tem que estar nessa ordem.

Podemos ter tantos elif's quanto precisar. Mas se usar essa estrutura, somente um único else é permitido

````
# Exemplo usando if/elif/else
nota=float(input('Qual a nota da prova?'))

if(nota<3):
  print('Reprovado com louvor')
elif(nota<5):
  print('Só está reprovado')
elif(nota<7.5):
  print('vai ter que fazer uma provinha')
else:
  print('Feitooo... passou!')

````

## Controle de Repetição - while

Sabemos que a repetição é um ítem importantíssimo de qualquer linguagem de programação. Há problemas computaionais que necessitam de que executem processos repetidamente. Ou **enquanto** uma condição for satisfeita.

Sintaxe:

```
while condição:
  comandos a serem executados

```

os comandos de dentro do **while** serão executados **enquanto** a condição for verdadeira.

````
x=0 # inicializo um contador para que controle o numero de vezes o bloco
    # while será executado
while (x<10):  # enquanto o contador acima for menor que o numero de vezes...
  print (x)    # imprimo (ou faço qualquer outra coisa)
  x=x+1        # incremento o contador. Senão o while fica "travado no zero"
````

## *break* e *continue*
Quando se trabalha com laços de repetição poderá haver situações em que precisamos "pular" o resto do programa, ou indo para a próxima interação ou saindo do laço.

O comando **break** faz com que o laço seja interrompido. Independente da condição vigente no comando while.

O comando **continue** faz com que o resto do corpo do while nao seja executado, saltando diretamente para o topo do comando while indo para a próxima interação.

Vamos para um exemplo legal: A Conjectura de Collatz.

É um problema matemático em que funciona assim. Se um número é par, divida-o por 2; se for ímpar, multiplique por 3 e adicione 1. Faça isso para sempre. O resultado sempre será 1.
image.png

Interessante, não? Dizem que tem um prêmio de US$1.000.000,00 para quem achar um número que não obedeça a essa conjectura!

````
# Conjectura de Collatz!

num=0
while(num<=0):
  num=int(input("Um numero qualquer, inteiro e positivo:"))

while True:
  print(num)

  if(num==1):
    break; # paro o programa

  if (num % 2)==0: # é par
    num=num//2 # divisão inteira, mas divido por 2
    continue # pulo para o while True...

  # é impar
  num=3*num+1

print("FIM :)")
````
